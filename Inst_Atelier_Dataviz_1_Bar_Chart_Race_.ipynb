````
# R√©ulitisez le code de la cellule ci-dessus et stockez le nom des pays dans une liste appel√©e 'countries_name' et les urls des drapeaux dans une liste appel√©e 'countries_flag'
containers = html_soup.find_all('div', class_='thumb')
countries_name=[]
countries_flag=[]
for container in containers: 
  try:
    country_name = container.find('span', class_='title').text[8:]
    country_flag_url = container.find('img')['src']
    #print(country_name, end=" ")
    #print(country_flag_url)
    sleep(0.05)
    countries_name.append(country_name)
    countries_flag.append(country_flag_url)
  except:
    pass
print(countries_name)
print(countries_flag)
````

````
# Cr√©ez un dataframe 'df_countries_flag' √† partir de nos listes 'countries_name' et 'countries_flag'.
df_countries_flag = pd.DataFrame(countries_flag, countries_name)
df_countries_flag = df_countries_flag.reset_index()
df_countries_flag = df_countries_flag.rename(columns={'index' : 'country_name', 0 : 'country_flag'})
df_countries_flag
````

### Feature Engineering

````
# Le code suivant v√©rifie si le nom d'un pays est pr√©sent √† la fois dans le dataset 'pop_df' et 'df_countries_flag'. Si c'est le cas, alors l'url
# du drapeau du pays est transf√©r√© dans le dataset 'pop_df', dans une nouvelle colonne appel√©e 'url_flag'. Dans le cas contraire, la valeur -1 sera
# ajout√©e. Le code applique cela √† tous les pays du dataset 'pop_df'.

pop_df['url_flag'] = (pop_df.country.apply(lambda x: df_countries_flag.query('@x in country_name').index.values)
                         .apply(lambda x: x[0] if len(x) else -1)).apply(lambda x: df_countries_flag.country_flag[x] if x != -1 else x)
````

````
# importez la librairie pyconvert-convert ici.
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
````

<i>Aide: Une fois les codes continents de chaque pays r√©cup√©r√©s et ins√©r√©s dans la colonne <b>region</b> de notre dataset <b>pop_df</b>, les convertir en <b>4 regions</b> (cf. instructions ci-dessus). Nous utiliserons le nom complet des r√©gions afin de les utiliser en l√©gende de notre visualisation.</i>

````
# Extraire la colonne 'country' depuis notre dataframe et la transformer en liste
list_countries = pop_df['country'].tolist()
list_countries
````

````
# Utilisez la librairie pycountry_convert afin d'associer chaque pays √† son code continent. Indice: nous utiliserons ici le format 'alpha2' pour obtenir
# de meilleurs r√©sultats.
# Afin d'√©viter tout arr√™t d'ex√©cution de notre code en cas de "non match", utilisez les instructions (statements) 'try except'.
NaN = float("NaN")
list_code_continents = []
for i in list_countries:
  try: 
    cn_a2_code =  country_name_to_country_alpha2(i)
  except:
    cn_a2_code = NaN
  try:
    cn_continent = country_alpha2_to_continent_code(cn_a2_code)
  except:
    cn_continent = NaN
  list_code_continents.append(cn_continent)
````

````
# Cr√©ez un nouveau dataframe qui contiendra les noms des pays et codes continents
# Pour cela, la fonction zip() pourra vous √™tre utile. Documentez-vous pour comprendre comment l'utiliser.
df_code = pd.DataFrame(list(zip(list_countries, list_code_continents)))
df_code = df_code.rename(columns={0 : 'country_name', 1 : 'code_continent'})
df_code.isna().sum()
````

````
# Trouver une m√©thode avec pandas afin d'extraire la liste des continents UNIQUES de notre nouveau dataframe.
# Il y a t-il des valeurs non renseign√©es ('NaN')?? Si oui, notre prochaine mission sera de les transformer.
df_code['code_continent'].unique()
````

````
# Faire un filtre (masque) afin de trouver toutes les lignes de notre dataframe contenant des NaN dans la colonne des codes continents.
mask = df_code['code_continent'].isna()
df_code[mask]
index = df_code[mask].index.values
df_code.iloc[2]['code_continent']
````

````
# Il y a peu de 'NaN'. Nous allons donc pouvoir les remplacer manuellement dans notre liste 'continent_code'.
# Pour cela, recup√©rez le num√©ro de ligne (index) de notre dataframe pour les codes continents ayant pour valeur 'NaN'.
# √Ä partir de ces index, remplacer les 'NaN' de notre liste 'continent_code' par les √©l√©ments de la liste ci-dessous:

continents_to_add = ['AF', 'AF', 'AF', 'EU', 'AS', 'OC', 'NA', 'AS']
for i in range(len(index)):
  df_code.iloc[index[i]]['code_continent'] = continents_to_add[i]
````

````
# Une fois la manipulation faite, reste t-il des 'NaN' dans notre liste?
df_code['code_continent'].unique()
````

````
# Ins√©rez la nouvelle liste 'continent_code' en 2√®me colonne de notre dataframe 'pop_df'.
# Pour cela, vous pourrez utiliser la m√©thode '.insert()'
pop_df.insert(1,'code_continent',df_code['code_continent'])
````

````
pop_df = pop_df.rename(columns={'code_continent' : 'region'})
````

````
# Afin de limiter le nombre le labels dans notre visualisation, transformez les codes continents en 4 r√©gions du monde 
# ('Asia', 'Europe', 'Africa', 'Americas'). Pour cela, comprenez et ex√©cutez le code suivant:

pop_df.loc[pop_df['region'] == 'NA', 'region'] = 'AM'
pop_df.loc[pop_df['region'] == 'SA', 'region'] = 'AM'
pop_df.loc[pop_df['region'] == 'OC', 'region'] = 'AS'

pop_df.loc[pop_df['region'] == 'AS', 'region'] = 'Asia'
pop_df.loc[pop_df['region'] == 'EU', 'region'] = 'Europe'
pop_df.loc[pop_df['region'] == 'AF', 'region'] = 'Africa'
pop_df.loc[pop_df['region'] == 'AM', 'region'] = 'Americas'
````
